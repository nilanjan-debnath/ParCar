```python
import pickle
import json
```

```python
try:
    with open('CarParkPos', 'rb') as f:
        posList = pickle.load(f)
except:
    posList = []
print(len(posList))
for i in posList:
    print(i)
```

```python
type(posList[1])
```

```python

details_list = []
```

```python
i = 1

for loc in posList:
    details = {
    "Slot_No": i,
    "Position": loc,
    "State": 0,
    "Name": "",
    "Booking_Time": "",
    "Leaving_Time": "",
    "Car_No": ""
    }
    details_list.append(details)
    i += 1

    
```

```python
details_list
```

```python
data = {"Parking_Spots": details_list}
```

```python
with open("sample2.json", "w") as outfile:
    json.dump(data, outfile)
```
